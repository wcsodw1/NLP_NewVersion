```python
#將斷好詞的list "article_cutted" load 進來  (load 'article_cutted')
#os.chdir("Part2/data")
#os.getcwd()

with open('data/article_cutted', 'rb') as file:
    data = pickle.load(file)
```

## build word2vec
   -  size=256 : 將字化成多少維度的詞向量
   -  min_count=5 :
   -  word2vec的 window 要設為多大
   -  sg=0 CBOW ; sg=1 skip-gram
     

```python
# build word2vec
# sg=0 CBOW ; sg=1 skip-gram
model = word2vec.Word2Vec(size=256, min_count=5, window=5, sg=0)
```

- 建立id對vocabulary的mapping

```python
# build vocabulary
model.build_vocab(data)
```

```python
# train word2vec model ; shuffle data every epoch
for i in range(20):
    
    #依據經驗 在training前做一個shuffle效果會比較好
    #shuffle() 方法将序列的所有元素随机排序
    random.shuffle(data)
    model.train(data, total_examples=len(data), epochs=1)
```

```python
## print an example
model.wv['人工智慧']
```

```python
## save model
#model.save('word2vec_model/CBOW')

model.save('word2vec_model.CBOW')
```

- ##  Try skip-gram

```python
## 嘗試調整 word2vec 的參數 , etc. sg = 1 = skip-gram
modelskip_gram = word2vec.Word2Vec(size=50, min_count=20, window=20, sg=1)
```

```python
# build vocabulary
modelskip_gram.build_vocab(data)
```

```python
# train word2vec model ; shuffle data every epoch
for i in range(10):
    random.shuffle(data)
    modelskip_gram.train(data, total_examples=len(data), epochs=1)
```

```python
## print an example
modelskip_gram.wv['人工智慧']
```

```python
## save model
modelskip_gram.save('word2vec_model.skip-gram')
```

# the application of word2vec

```python
## load word2vec model

#os.chdir("Part2/")

model = word2vec.Word2Vec.load('word2vec_model.CBOW')
```

## clustering

```python
## create a dictionary: words as key ; count as values
# 同字詞出現次數

words = {word: vocab.count for word, vocab in model.wv.vocab.items()}
words 
```

```python
diction = {'Google': 'www.google.com', 'Runoob': 'www.runoob.com', 'taobao': 'www.taobao.com'}
 
print("字典值 : %s" %  diction.items())
```

```python
for i, (x,y) in enumerate(words.items()):
    print(i,"  ",x,"  ",y)
    if i >=10: break
```

```python
## sort and select the top 10000 count of words
words = sorted(words.items(), key=lambda x: x[1], reverse=True)
words = words[:10000]
words = np.array(words)[:, 0]
words
```

```python
## extract the word vectors 
vecs = model.wv[words]
type(vecs) #numpy array
vecs
```

- ## k means 分群

```python
## run clustering algorithm
kmeans = KMeans(n_clusters=50)
cluster = kmeans.fit_predict(vecs)
```

```python
print(kmeans)
```

```python
print(cluster)
```

```python
## print the result
df = pd.DataFrame([words.tolist(), cluster.tolist()], index=['words', 'no. cluster']).T
df.head(n=50)
```

```python
## print every cluster of words

data = pd.concat([d['words'].reset_index(drop=True).rename(str(k)) for k, d in df.groupby('no. cluster')],axis=1)
data
```
